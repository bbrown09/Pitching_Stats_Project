```python
import numpy as np
import requests
import os
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#I needed to scrape 8 total pages of stats, so for the first page I seperated my steps so the code would be easier to follow
#but for the rest of the pages I put all of the code in the same cell to speed up the process.

url = "https://www.mlb.com/stats/pitching/games/2021"

page = requests.get(url)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
#This shows that the code is fair game for us to scrape legally which makes sense since it is a public sports database.

page.status_code
```

```python
#All of the stats are under the table tag

table1 = soup.find("table", class_="bui-table is-desktop-sKqjv9Sb")
```

```python
#All of the headers were collected here to form the outline of our data

headers = []
for i in table1.find_all("th", scope="col"):
 title = i.text
 headers.append(title)
del headers[0]
```

```python
#For some reason the names of the players were seperated from the other stats so we make a dataframe here that we can just
#merge with the other stats dataframe.

name = []
for i in table1.find_all("th", scope="row"):
 title = i.text
 name.append(title)
names = pd.DataFrame(name)
```

```python
mydata = pd.DataFrame(columns = headers)
```

```python
#Here we fill our dataframe with the actual numbers

for j in table1.find_all("tr")[1:]:
 row_data = j.find_all("td")
 row = [i.text for i in row_data]
 length = len(mydata)
 mydata.loc[length] = row
```

```python
#Here we merge the names and the stats together

frames = [names, mydata]

result1 = pd.concat(frames, axis=1)
result1
```
