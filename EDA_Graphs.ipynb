```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv('2013-21PitchingStats.csv')
df1
```

```python
#Since we filtered by pitchers with the most games, none of them played complete games or had shutouts
df1.drop('CG', inplace=True, axis=1)
df1.drop('SHO', inplace=True, axis=1)
df = df1
```

```python
#Checking for any interesting correlations
#Most make sense here
sns.set(rc={"figure.figsize":(15, 10)})
sns.heatmap(df.corr(), annot=True)
```

```python
sns.scatterplot(data=df, x='IP', y='HR', style='Year', s=100)
```

```python
sns.regplot(data=df, x='Year', y='HR')
plt.title('Top 25 Pitchers Homeruns By Year', fontsize= 18)
plt.savefig('Homeruns_By_Year.png', bbox_inches = 'tight', pad_inches = 1)
```

```python
sns.regplot(data=df, x='Year', y='H')
plt.title('Top 25 Pitchers Hits Allowed By Year', fontsize= 18)
plt.savefig('Hits_By_Year.png', bbox_inches = 'tight', pad_inches = 1)
```

```python
sns.regplot(data=df, x='Year', y='ERA')
plt.title('Top 25 Pitchers ERA By Year', fontsize= 18)
plt.savefig('ERA_By_Year.png', bbox_inches = 'tight', pad_inches = 1)
```

```python
sns.regplot(data=df, x='Year', y='SO')
plt.title('Top 25 Pitchers Strikeouts By Year', fontsize= 18)
plt.savefig('Strikeouts_By_Year.png', bbox_inches = 'tight', pad_inches = 1)
```

```python
sns.scatterplot(data=df, x='Year', y='HR', hue='ERA', palette="Greens")
```

```python
sns.scatterplot(data=df, x='SV', y='SVO', hue='SV')
plt.title('Saves Compared to Save Opportunities', fontsize= 18)
plt.savefig('Saves_Vs_Save_Opportunities.png', bbox_inches = 'tight', pad_inches = 1)
```

```python
sns.boxplot(data=df, x='W', y='HR')
plt.title('Homeruns Compared to Wins in a Year', fontsize= 18)
plt.savefig('Homeruns_Compared_to_Wins.png', bbox_inches = 'tight', pad_inches = 1)
```

```python
sns.regplot(data=df, x='SV', y='HR')
plt.title('Homeruns Compared to Saves in a Year', fontsize= 18)
plt.savefig('Homeruns_Compared_to_Saves.png', bbox_inches = 'tight', pad_inches = 1)
```
